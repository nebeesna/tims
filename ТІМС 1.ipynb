```python
def get_Me(values, counts = None):
    if type(counts) == type(None):
        if len(values)%2 == 1:
            return values[int(len(values)/2)]
        else:
            return (values[int(len(values)/2)] + values[int(len(values)/2) - 1])/2
    else:
        index = 0
        if sum(counts)%2 == 1:
            find = int(sum(counts)/2)
            for i in range(len(values)):
                index += counts[i]
                if index >=find:
                    return values[i]
        else:
            find = int(sum(counts)/2)
            for i in range(len(values)):
                index += counts[i]
                if index > find:
                    return values[i]
                elif index==find:
                    return (values[i] + values[i+1])/2
```

```python
def get_Mo(values, counts = None):
    if type(counts) == type(None):    
        values, counts = np.unique(values, return_counts=True)
    max_ = max(counts)
    res = []
    for i in range(len(counts)):
        if counts[i] == max_:
            res.append(values[i])
    return res
```

```python
def get_average(values, counts = None):
    if type(counts) == type(None):
        return sum(values)/len(values)
    return sum(values*counts)/sum(counts)
```

```python
def get_dev(values, counts = None):
    arr_res = []
    if type(counts) == type(None):
        arithmetic_mean = get_average(values)
        for xi in values:
            arr_res.append((xi - arithmetic_mean)**2)
    else:
        arithmetic_mean = get_average(values, counts)
        for i in range(len(values)):
            arr_res.append(((values[i] - arithmetic_mean)**2 )*counts[i])
    return sum(arr_res)
```
