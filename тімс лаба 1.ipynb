```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# **Статистичний матеріал**

```python
statistical_material = np.random.randint(0,50,100)
print(statistical_material)
```

# **Варіаційний ряд**

```python
variation_range = np.array(sorted(statistical_material))
print(variation_range)
```

# **Частотна табличка** 

```python
unique, count = np.unique(variation_range,return_counts = True)
frequency_table = pd.DataFrame({"unique":unique, "count":count})
frequency_table
```

```python
print(f"Перевірка кількості: {count.sum()}")
```

```python
plt.title("Діаграма частот")
plt.bar(frequency_table['unique'],frequency_table['count'], width=0.5)
plt.grid(1)
plt.show()
```

```python
plt.title("Полігон частот")
plt.plot(frequency_table['unique'],frequency_table['count'], marker="o")
plt.grid(1)
plt.show()
```

# **Функція розподілу**

```python
print(f"0,\t x<{unique[0]}")
sum = frequency_table['count'].sum()
probability = 0
probability_list = []
probability_list.append(0)
for i in range(count.size):
    probability += count[i]/sum
    probability_list.append(probability)
    if i == count.size - 1:
        print(f"{probability.round(6) },\t {unique[i]} <= x")
    else:
        print(f"{probability.round(6) },\t {unique[i]} <= x < {unique[i+1]}")
```

```python
plt.grid(1)
plt.plot([unique[0]-5, unique[0]], [probability_list[0], probability_list[0]])
for i in range(1, len(probability_list) -1):
    plt.plot([unique[i]-1, unique[i]],[probability_list[i], probability_list[i]])
plt.plot([unique[-1], unique[-1]+5], [probability_list[-1], probability_list[-1]])
plt.show()
```

# **Статистика локації**

```python
n = count.sum()
print(f"n = {n}")
max_count = count.max()
mode = []
for i in range(len(count)):
    if count[i] == max_count :
        mode.append(unique[i])
print(f"Мода = {mode}")

if unique.size%2 == 0:
    print(f"Meдіана = {(unique[int(unique.size/2)] + unique[int(unique.size/2)+1])/2}")
else:
    print(f"Meдіана = {unique[int((unique.size-1)/2)]}")

x = variation_range.sum() / n
print(f"Середнє вибіркове = {x}")
d = 0
for i in unique:
    d += (i - x)**2
    
print(f"Девіація = {d}") 
```

# **Статистика розсіяння**

```python
s = (d / (n-1))**(1/2)
v = s/x
D = d / n
print(f"Варіанса = {s**2}")
print(f"Розмах = {unique.max() - unique.min()}")
print(f"Стандарт = {s}")
print(f"Варіація = {v}")
print(f"Дисперсія = {D}")
```

```python
Q = [variation_range[int(n/4)-1], variation_range[int(2*n/4)-1], variation_range[int(3*n/4)-1]]
print(f"Квартилі: {Q}")
print(f"Інтерквартильна широта: {Q[2] - Q[0]}")
Oct = []
for i in range(1,8):
    Oct.append(variation_range[int(i*n/8)-1])
Dec = []
for i in range(1,10):
    Dec.append(variation_range[int(i*n/10)-1])
print(f"Децилі: {Dec}")
print(f"Інтердецильна широта: {Dec[8] - Dec[0]}")
```

```python
def get_moment(var_range, k, a = 0):
    m = 0
    for i in var_range:
        m += (i - a)**k
    m /= len(var_range)
    return m
```

```python
print("Початкові моменти поряку")
print(f"Першого: {get_moment(variation_range,1)}")
print(f"Другого: {get_moment(variation_range,2)}")
print(f"Третього: {get_moment(variation_range,3)}")
print(f"Четвертого: {get_moment(variation_range,4)}")
```

```python
print("Центральні моменти поряку")
print(f"Першого: {round(get_moment(variation_range,1, x),10)}")
print(f"Другого: {get_moment(variation_range,2, x)}")
print(f"Третього: {get_moment(variation_range,3, x)}")
print(f"Четвертого: {get_moment(variation_range,4, x)}")
```

# **Статистики форми**

```python
A = get_moment(variation_range,3, x) /(get_moment(variation_range,2, x))**(3/2)
print(f"Асиметрія: {A}")
if A == 0:
    print("статистичний матеріал симетричний")
elif A > 0:
    print("статистичний матеріал скошений вправо і має додатну асиметрію")
elif A < 0:
    print("статистичний матеріал скошений вправо і має додатну асиметрію")

E = (get_moment(variation_range,4, x) /(get_moment(variation_range,2, x))**2) - 3
print(f"Ексцес: {E}")
if E == 0:
    print("статистичний матеріал нормальновершинний")
elif E > 0:
    print("статистичний матеріал високовершинний")
elif E < 0:
    print("статистичний матеріал низьковершинний")
```
